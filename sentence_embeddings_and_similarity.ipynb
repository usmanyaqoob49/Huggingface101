```python
from sentence_transformers import SentenceTransformer
from sentence_transformers import util
```

### Embeddings Model:

```python
model= SentenceTransformer("all-MiniLM-L6-v2")
```

```python
sentences= ['Computer is a useful machine.',
            'Laptop uses battery for usage without electriciyt.',
            'The dog is faithful animal']
```

```python
embeddings= model.encode(sentences,
                         #we want to get tensors
                        convert_to_tensor= True)

embeddings
```

```python
embeddings.shape
```

**Comparing sentences that are related (first and second)**

```python
cosine_scores1_2= util.cos_sim(embeddings[0],
                            embeddings[1])

cosine_scores1_2
```

**Now lets compare sentences that are not related**

```python
cosine_scores1_3= util.cos_sim(embeddings[0],
                            embeddings[2])

cosine_scores1_3
```
