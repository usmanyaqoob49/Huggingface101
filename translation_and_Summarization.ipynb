```python
from transformers import pipeline
import torch
```

# Translation of Language:

We will be use model from meta which can translate 200 different languages:

```python
translator= pipeline(task= 'translation',
                     model= 'facebook/nllb-200-distilled-600M',
                     #to compress the model without performance degradation
                     torch_dtype= torch.bfloat16)
```

Text to be translated:

```python
text= """
        I am ver shocked with your behaviour.
"""
```

```python
text_transltor= translator(text,
                        src_lang= 'eng_Latn',
                        tgt_lang= 'urd_Arab' )
```
